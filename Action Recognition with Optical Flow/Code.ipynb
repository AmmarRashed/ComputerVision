```python
x = np.arange(2, 6)
y = np.arange(2, 6)
z = np.array([[0. for i in range(len(x))] for j in range(len(y))])
for i in x:
    for j in y:
        z[j - 2][i - 2] = d[(i, j)]
```

```python
fig = pylab.figure(figsize=(12,7))
ax = Axes3D(fig)
ax.set_xlabel("B")
pylab.xticks(np.arange(2, 6))
ax.set_ylabel("K")
ax.set_zlabel("Accuracy")
pylab.yticks(np.arange(2, 6))
ax.plot_surface(x,y,z, rstride=1, cstride=1)
# plt.figure(figsize=(30,20))
plt.show()
```
